```python
import geopandas as gpd
```

```python
#list avilable datasets 
# naturalearth_cities - cities list 
# naturalearth_lowres - world boundries 
# nybb - New York boroughs
gpd.datasets.available
```

```python
#get crude world and country boundries 
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
world.head()
```

```python
#plot world plot static 
world.plot(figsize=(10,5))
```

```python
# plot unites states, similarly we can draw continents etc, boundries can collapsed as well  
world[world.name == "United States of America"].plot()
```

```python
# country population 
world.plot(column='pop_est', legend=True, figsize=(18,10))
```

```python
# Dynamic map using folium 
world.explore(column='pop_est',cmap='Set2')
```

```python
asia = world[world['continent'] =='Asia']
asia.explore(column='gdp_md_est',
             cmap='Set2',
             legend=False,
             tooltip=['name','gdp_md_est'])
```
