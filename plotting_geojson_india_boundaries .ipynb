```python
import folium

import requests as re
```

```python
urlData = re.get(f'https://raw.githubusercontent.com/datameet/maps/master/Country/india-composite.geojson').json()
urlData["features"][0]['geometry']['coordinates']
```

```python
boundary_folium = []
for pt in urlData["features"][0]['geometry']['coordinates'][0][0]:
    boundary_folium.append([pt[1], pt[0]])   
```

```python
map_osm = folium.Map(location=[22.55353, 88.24062], zoom_start=5, width=800, height=800)
folium.PolyLine(boundary_folium, color='black').add_to(map_osm)
map_osm
```
